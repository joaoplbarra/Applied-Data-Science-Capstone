<h2>Introduction/Bussiness</h2>

It is no secret that car accidents are a major cause of death and high repair costs for the city. Measures to remedy the number of accidents, fatal or not, must and are being taken. One of the measures is with the use of Machine Learning attributes, in which the algorithms and the vast database can draw different insights and conclusions regarding possible measures to be taken.

We can mention several factors involved in a car accident, such as car model, climate, climatic conditions, location, time until the driver's age or whether there were fatal victims or not. The biggest benefit of machine learning is being able to find correlations and patterns between the various factors that are involved in a car accident.

The target audience for this analysis is the Seattle Department of Transportation and Government. It must identify the main causes of accidents and allow them to identify trends when accidents can be avoided. This will reduce the number of accidents and injuries in the city.
